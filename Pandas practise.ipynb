### iloc:

### loc:

```python
df.loc[0]        # same as df.iloc[3]
df.loc[3,:]      # same as df.iloc[3,:] or df.loc[3]
df.loc[0:3]      # same as df.iloc[0:3]
#df.loc[:,3:6] ------- not same as df.iloc[:,3:6]
#df.loc[3,1] --------- not same as df.iloc[3,1]
#df.loc[0:3,4:7] ----- not same as df.iloc[0:3,4:7]
#df.loc[0:3,4:] ------ not same as df.iloc[0:3,4:]

df.loc[0,'Type 1']                     # at row_label: 0, column_label: Type 1
df.loc[0:3, ['Type 1','HP','Attack']]  # from row_label 0-3 and selected column label
df.loc[0:3, 'Type 1':]                 # from row_label 0-3 and from column_label 'Type 1' to last
df.loc[0:3, 'Type 1':'Defense']        # from row_label 0-3 and from column_label 'Type 1' to 'Defense'
df.loc[[0,1,4], ['Type 1','Defense']]  # only row_labels 0,1,4 and column_label 'Type 1','Defense' --- eqv to iloc
```

```python
df.loc[[2,5],['HP']]==30      # boolean result for row: 2,5 for column_label: HP, its False here
df.loc[2:5,['HP']]==30        # boolean result from row_label: 2-5 for column_label: HP, its False here
#df.loc['Name']=='Charmander' ---- won't work
df.loc[4]=='Charmander'       # boolean check for row_label:4 and all column at that index with 'Charmander'
```

```python
# Difference between df.loc[...] AND df.loc[df[..]<condition>] is that: 
#        former is boolean check AND latter is display

df.loc[df['HP']<60]                # data display with HPless than 60
#df.loc[df['HP','Attack']<80]      # data display with HPless than 60 (not working)
df.loc[df['Type 1']=='Grass']      # data display with Type 1 has only Grass
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['Defense']>60)]   # multiple conditions with loc
```
